Завдання 8. Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні
від 1 до 8 та знайдіть її обернену матрицю.


```python
import numpy as np

# Створення матриці 2x2 з випадковими цілими числами від 1 до 8
matrix = np.random.randint(1, 9, size=(2, 2))

# Виведення значень матриці
print("Матриця:\n", matrix)

# Перевірка на наявність оберненої матриці (матриця повинна бути невиродженою)
if np.linalg.det(matrix) != 0:
    # Обчислення оберненої матриці
    inverse_matrix = np.linalg.inv(matrix)
    print("Обернена матриця:\n", inverse_matrix)
else:
    print("Матриця є виродженою, обернена матриця не існує.")
```

Завдання 15. Створіть матрицю розміром 5x5, заповніть її випадковими дійсними числами в
діапазоні від -0.5 до 0.5 та знайдіть вектор-стовпчик, що містить суму елементів
кожного рядка матриці.


```python
import numpy as np

# Створення матриці 5x5 з випадковими дійсними числами від -0.5 до 0.5
matrix = np.random.uniform(-0.5, 0.5, size=(5, 5))

# Виведення значень матриці
print("Матриця:\n", matrix)

# Обчислення суми елементів кожного рядка
row_sums = np.sum(matrix, axis=1)

# Перетворення на вектор-стовпчик
row_sums_column_vector = row_sums.reshape(-1, 1)

# Виведення результату
print("Вектор-стовпчик сум елементів кожного рядка:\n", row_sums_column_vector)
```
