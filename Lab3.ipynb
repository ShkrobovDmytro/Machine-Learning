```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import TargetEncoder, StandardScaler, LabelEncoder

import torch
import torch.nn as nn
import torch.optim as optim

import warnings
warnings.filterwarnings('ignore')
```

```python
X = df.drop('Cured', axis=1).values
y = df['Cured'].values
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42, test_size=0.2, stratify=y)
```

```python
X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)

X_test = torch.tensor(X_test, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32).unsqueeze(1)
```

```python
m = nn.Linear(13,1)
input = torch.randn(1, 13)
output = m(input)

print('Input:', input, f'shape {input.shape}', sep='\n')
print('\nOutput:', output, f'shape {output.shape}', sep='\n')
```

```python
t = torch.randn(4)
print('Input; ', t)
print('Applying sigmoid: ', torch.sigmoid(t))
```

```python
class LogisticRegression(nn.Module) :
  def __init__(self, input_dim):
    super(LogisticRegression, self).__init__()
    self.linear = nn.Linear(input_dim, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, x) :
    out = self.linear(x)
    out = self.sigmoid(out)
    return out
```

```python
input_dim = X_train.shape[1]
model = LogisticRegression(input_dim)

criterion = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
model.parameters
```

```python
num_epochs = 500
for epoch in range(num_epochs):
  outputs = model(X_train)
  loss = criterion(outputs, y_train)

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  if (epoch +1) % 5 == 0:
    print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')
```

```python
with torch.no_grad():
  y_pred = model(X_test).squeeze().numpy()
  y_pred = (y_pred > 0.5).astype(int)
  accuracy = accuracy_score(y_test.numpy(), y_pred)
  print('Accuracy:', accuracy)
```

### Висновок

У ході виконання лабораторної роботи було застосовано три методи кластеризації: метод k-середніх (KMeans), ієрархічну кластеризацію та DBSCAN.

Методи KMeans та ієрархічної кластеризації показали схожі результати, обидва виділили 3 основні кластери. Однак у KMeans спостерігалося чіткіше розділення між кластерами, що пояснюється його припущенням про сферичну форму кластерів. Ієрархічна кластеризація враховувала структуру даних на різних рівнях, проте могла давати менш чітке розділення.

Метод DBSCAN продемонстрував інший підхід до кластеризації — він орієнтується на густину точок. У результаті було виділено лише 2 кластери, а частину точок віднесено до шуму. Це може бути корисним у випадках, коли в даних присутні викиди або неоднорідна густина розподілу.

Отже, вибір методу кластеризації залежить від структури даних та завдань аналізу: KMeans добре працює з чітко розділеними кластерами, ієрархічний підхід — для дослідження вкладеної структури, а DBSCAN — для виявлення кластерів довільної форми та шумів.
